```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img=cv.imread("C:\\Users\\leonf\Music\\image_processing\\emma.jpg",1)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
fig,im=plt.subplots(1,2,figsize=(7,7))
arr1=np.linspace(0,50,51).astype(np.uint8)
arr2=np.linspace(101,255,100).astype(np.uint8)
arr3=np.linspace(150,255,105).astype(np.uint8)
arr=np.concatenate((arr1,arr2))
arr=np.concatenate((arr,arr3))
img_new=cv.LUT(img,arr)
im[0].imshow(img)
im[0].set_title("Without transformation")
im[1].imshow(img_new)
im[1].set_title("with transformation")
```

```python
img=cv.imread("C:\\Users\\leonf\Music\\image_processing\\BrainProtonDensitySlice9.png",0)
white_arr1=np.arange(255,-1,-1)
arr1=np.ones((1,50))*255
arr1=arr1.squeeze().astype(np.uint8)
arr2=np.arange(155,-1,-1).astype(np.uint8)
white_arr=np.concatenate((arr1,arr2))
white_img=cv.LUT(img,white_arr)

arr3=np.arange(0,156,1).astype(np.uint8)
black_arr=np.concatenate((arr3,arr1))
#print(f"{len(arr1)+len(arr3)}")
black_img=cv.LUT(img,black_arr)
fig,im=plt.subplots(1,3,figsize=(8,8))
im[0].imshow(img,cmap="gray",vmin=0,vmax=255)
im[0].set_title("Without transformation")
im[1].imshow(white_img,cmap="gray",vmin=0,vmax=255)
im[1].set_title("with transformation")
im[2].imshow(black_img,cmap="gray",vmin=0,vmax=255)
im[2].set_title("with transformation")

```

```python
img=cv.imread("C:\\Users\\leonf\Music\\image_processing\\highlights_and_shadows.jpg",1)
img=cv.cvtColor(img,cv.COLOR_BGR2LAB)
img_new=np.copy(img)
gamma=1.3
img_new[:,:,0]=(((img[:,:,0]/255)**gamma)*255).astype(np.uint8)
fig,ax=plt.subplots(1,2,figsize=(15,18))
img=cv.cvtColor(img,cv.COLOR_LAB2RGB)
img_new=cv.cvtColor(img_new,cv.COLOR_LAB2RGB)
ax[0].imshow(img)
ax[1].imshow(img_new)
```

```python
color=['b','g','r']
fig,ax=plt.subplots(1,2,figsize=(10,5))

for i,c in enumerate(color):
    hist=cv.calcHist([img],[i],None,[256],[0,256]) 
    ax[0].plot(hist,color=c) 
    ax[0].set_xlim([0,256])

color=['b','g','r']
for i,c in enumerate(color):
    hist=cv.calcHist([img_new],[i],None,[256],[0,256]) 
    ax[1].plot(hist,color=c) 
    ax[1].set_xlim([0,256])

plt.tight_layout()
plt.show()
```
